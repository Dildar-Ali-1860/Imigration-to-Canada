```python
#Comparing the trend of top 5 countries that contributed the most to immigration to Canada.

inplace=True
data.sort_values(by='Total',ascending=False,inplace=True)

top5_data = data.head(5)
top5_data
```

```python
data_top = top5_data[years].transpose()
data_top
```

```python
data_top.plot(kind='line',figsize=(14,6))
plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')


plt.show()
```
